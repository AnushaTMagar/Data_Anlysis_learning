```python
# filtering videos with "claim" status
claim_videos = df[df['claim_status'] == 'claim']
```

```python
# average view count
mean_view_count = claim_videos['video_view_count'].mean()
median_view_count = claim_videos['video_view_count'].median()
print(f"Mean view count of videos with 'claim' status is : {mean_view_count}")
print(f"Median view count of videos with 'claim' status is : {mean_view_count}")
```

### What is the average view count of videos with "opinion" status?

```python
# filtering videos with "opinion" status
opinion_video = df[df['claim_status']=='opinion']
```

```python
# average view count with 'opinion'
mean_view_counts_opinion = opinion_video['video_view_count'].mean()
median_view_counts_opinion = opinion_video['video_view_count'].median()
print(f"The Mean view count of video with 'opinion' status is: {mean_view_counts_opinion}")
print(f"The Median view count of video with 'opinion' status is: {median_view_counts_opinion}")
```

### Question: What do you notice about the mean and media within each claim category?


View counts for claim status videos are evenly distributes arount the average which suggest symmetric distributons and provide good representation of average view count while difference between mean and median for 'opinion' status suggests skewness towards higher view count. So, for 'opinion' status videos, the median might be a better measure of central tendency due to the skewness.

```python
# Get counts for each group combination of claim status and author ban status
df.groupby(['claim_status','author_ban_status']).count()[['#']]
```

#### Question: What do you notice about the number of claims videos with banned authors? Why might this relationship occur?

Based on the data from the table, we observe a higher number of banned authors for claim videos than for opinion videos. This could be due to several factors:

- Claim videos are subject to stricter monitoring than opinion videos.
- There are more stringent rules for authors when posting claim videos compared to opinion videos.
- Authors of claim videos may be more likely to post content that violates terms of service.  

In conclusion, the stricter oversight and regulations for claim videos likely contribute to the higher incidence of bans for their authors.  

#### Continue investigating engagement levels, now focusing on author_ban_status.

#### Calculate the median video share count of each author ban status.

```python
# median video share count of each author ban status
df.groupby(['author_ban_status']).agg(
        {'video_view_count': ['median'],
         'video_like_count': ['median'],
         'video_share_count':['median'] 
        })
```

```python
# What's the median video share count of each author ban status?
df.groupby(['author_ban_status']).median(numeric_only = True)[['video_share_count']]
```

```python
# exploring indepth of author_ban_status
df.groupby(['author_ban_status']).agg(
        {'video_view_count': ['count','median'],
         'video_like_count': ['count','median'],
         'video_share_count':['count','median'] 
        })
```

The data indicates that video from benned or under review  author receive much more engagement i.e. than that of active authors. This high engagement might attract more attention from the platform, leading to stricter monitoring and enforcement actions.

```python
# create a likes_per_view column
df['likes_per_view']= df['video_like_count']/df['video_view_count']

# create a comments_per_view column
df['comments_per_view'] = df['video_comment_count']/df['video_view_count']

# create a shares_per_view column
df['shares_per_view']= df['video_share_count']/df['video_view_count']
```

```python
df.groupby(['claim_status','author_ban_status']).agg(
        {'likes_per_view': ['count','median'],
         'comments_per_view': ['count','median'],
         'shares_per_view':['count','median'] 
        })
```

Videos from banned and under-review authors get more views, likes, and shares than those from non-banned authors. However, engagement rates (likes, comments, shares per view) depend more on the type of video (claim vs. opinion) than the author's ban status.

Claim videos have higher view, like, comment, and share rates than opinion videos, indicating they are more engaging and favorably received. For claim videos, banned authors have slightly higher engagement rates compared to active or under-review authors. In contrast, for opinion videos, active and under-review authors have higher engagement rates than banned authors.
