```python
import pandas as pd
import seaborn as sns
sns.set()
```

```python
cities = pd.read_csv('california_cities.csv')
cities.head()
```

```python
cities.shape
```

```python
alt, city = cities['Altitude'], cities['City']
area, population = cities['Area'], cities['Population']
# population, area = cities['Population'], cities['Area']
```

```python
plt.figure(figsize =(3,3),dpi=200)

#scatter plot the points using size and color but no label
plt.scatter(alt,population, label = None, c= np.log10(population),cmap= 'viridis', s= area, linewidth=0,alpha= 0.5)

plt.xlabel('Altitude')
plt.ylabel('City')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3,7)


# Let's create a legend
# we will plot empty lists with the desires size and label
for area in [100,300,500]:
    plt.scatter([], [], c= 'k', alpha = 0.3, s= area, label = str(area)+'km$^2$')
plt.legend(scatterpoints=1, frameon= False, labelspacing = 1, title = 'City Area')
plt.title('California CIties: Area and Population')
```
